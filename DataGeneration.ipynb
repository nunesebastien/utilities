```python
import random
import datetime

write = False

file1 = open("/Users/sebastiennunes/Documents/McGill/Schoolwork/COMP 421/sqlstatements.txt","a")#append mode 
file2 = open("/Users/sebastiennunes/Documents/McGill/Schoolwork/COMP 421/sqlstatementsORDER.txt","a")#append mode

users = [['sebnunes', 1], ['cjeon', 2], ['isabelledm', 3], ['nimachats', 4]]
couriers = ['dhl', 'canadapost', 'ups', 'fedex']
j = 0
#loop through the users
for user in users  :
    #each user has 30 orders
    h = 30;
    for i in range(1,30) :
        
        #generate a bunch of dates, each loop genereates a date a week earlier
        x = datetime.datetime.now() - datetime.timedelta(days = 7*i)
        date = str(x)[:10]
        
        #generate an order_id
        order_id = str(j*100 + i)
        
        #choose a random courier
        courier = couriers[random.randint(0,3)]
        if courier == 'dhl' :
            priority = 'express'
        else: priority = 'standard'
        
        #get shipping address id (really just a placeholder)
        address_id = str(user[1])
        
        #define status
        helper = random.randint(0,10)
        if helper == 10 :
            status = 'resolved'
        else : status = 'delivered'
        
        toWriteORDER = "INSERT INTO order VALUES ('{}', '{}', '{}', '{}', '{}', '{}', '{}', '{}');"
        if write == True :
            file2.write(toWriteORDER.format(order_id, str(user[0]), date, courier, status, str(0), address_id, priority))
            file2.write("\n")
        
        #this defines the "growth" in our business's sales (by restricting max items/order the futher back we go)
        nbItemsInOrder = random.randint(2,(h + 2))
        h -= 1
        #fill the order with items
        item_IDS = []
        for k in range (1, nbItemsInOrder) :
            number = random.randint(1,35)
            item_id = str(number)
            
            #ensure you are not adding any two of the same item to the order
            while item_id in item_IDS :
                number = random.randint(1,35)
                item_id = str(number)
                
            item_IDS.append(item_id)
            quantity = random.randint(1,20)
            
            toWritePurchasedITEM = "INSERT INTO purchasedItem VALUES ('{}', '{}', '{}', '{}');"
            if write == True :
                file1.write(toWritePurchasedITEM.format(order_id, str(user[0]), str(quantity), item_id))
                file1.write("\n")
            
    j += 1 
    
file1.close() 
file2.close()
```
